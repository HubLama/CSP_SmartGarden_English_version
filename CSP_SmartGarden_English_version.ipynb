##  **Arc Consistency(AC-3):**
  AC-3 one of the most well-known arc-consistency algorithms for filtering CSPs. This algorithm repeatedly carries out revisions and requires support checks for identifying and deleting all unsupported values from the domains.

 ### -AC-3 Algorithm Explanation:
 - Starting the algorithm by Turning each Binary Constraints into two arcs.
 -Then add all the arcs in a Queue.
 -Starting a repeated loop  until the queue is empty
	- dequeue the First arc (H1, H2)  from the queue.
	- send the arc to revise(arcs) method to make H1 arc consistent with H2.
 - The revise(x, y) method checks that the values(plants) of the x hole and y hole are satisfied the constraints and they are compatible. return True if the values of one of the holes are modified and False if nothing is deleted or changed.  
  
 -if the revised method did any change in the H1 domain, add all arcs with (x, H1).




 ### - AC-3 Implementation:



```python
import sys

def ac3(arcs,domain): #used to reduce the Domain size
    
    queue = arcs[:] # a queue with all CSP's arcs 
    
    while queue: #repeat while queue is not empty 
        (x, y) = queue.pop(0) #dequeue the First arc of the queue
        revised = revise(x, y,domain) #Method call to make x and y consistent 
        if revised: # if the method changes x's domain 
            neighbors = [neighbor for neighbor in arcs if neighbor[1] == x] #neighbors will have all the values matchs (?,x)
            queue = queue + neighbors # add neighbors to the queue

def revise(x, y,domain): # called by ac3 
    revised = False # if not modified domain x 
    x_domain = domain[x].copy() # to get x values(plants)
    y_domain = domain[y].copy() # to get y values(plants)
    
    if len(domain[y]) == 0:  
        sys.exit("No Solution")   # if there is no values in y 

    for x_value in x_domain:  # to check that the values of x satisfy constraints with y values
        c=0  # to ensure that it will only delete x_value if there is no values in y_domain can satisfy constraints
        for y_value in y_domain:
            if x_value in Constraints.keys():  # cheak if the x_value within Constraints
                if y_value not in Constraints[x_value]: 
                    break   # no need to continue the loop if at least one value can satisfy constraints
                else:
                    c+=1    # count the values not satisfy constraints
      
        if c == len(domain[y]): # if all values of y not satisfy constraints of x_value
            domain[x].remove(x_value) #delete x_value
            revised = True   # to indicate that the x_domain change 
   
    return revised
```

##  **Backtraking search:**
Backtracking search  is commonly used for solving CSPs, this algorithm looks for a complete consistent assignment to the variables.
### -  Backtracking search Algorithm Explanation:

- The algorithm begins by comparing the domains size with the assignment size, if both are equal the assignment will be returned, and if they are not equal will take these steps:
- All the values in domains variable will store in the unassigned variable.
- Then we stored the first value of the unassigned variable in the first variable.
- Then we starting repeated a loop on all the values in domains[first]:
    - In the repetition, we will use the consistent method to check the constraints on plants.
- If there is a solution the method will return an assignment for each variable in domains, or return None which means there is no solution.
	
### - Backtracking search Implementation:


```python

def backtracking_search(assignment):
    if len(assignment) == len(domains): # if the assignment size equal the domains size
        return assignment
  
    unassigned= [v for v in domains if v not in assignment] # get all variables in the domains but not in the assignment
  

    first= unassigned[0] # get the value of the first unassigned variable
    for value in domains[first]: # for loop on all the values of the first variable in domains
        local_assignment = assignment.copy() # all the values in assignment will store in local_assignment 
        local_assignment[first] = value # get the value from domains[first] variable
        if consistent(first, local_assignment): # if we're still consistent, we recurse (continue)  
            result=backtracking_search(local_assignment) 
            if result is not None: # if we find the result 
                return result
    return None
    
def consistent(x, assignment): # check if consistent 
         
    if len(assignment) == 1: # if the assignment contain one value  
        return True
    y=None
    #R=[('H1', 'H2'),('H2', 'H3')] 
    for neighbor in R: # to find the second value in the relation
        if neighbor[0] == x:
            y=neighbor[1]
            break
        if neighbor[1] == x: 
            y=neighbor[0]
            break
    y_value = assignment[y] # to get y values from the assignment 
    x_value = assignment[x] # to get x values from the assignment 
    if x_value in Constraints:
        if y_value in Constraints[x_value]: # check if Constraints[x_value] contain y_value
            return False
    
    return True 
```
